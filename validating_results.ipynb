```python
def get_preds(done_parts_folder_path):
    part_i = 1
    preds = []
    while True:
        try:
            done_part_file_path = done_parts_folder_path + f'preds_part_{part_i}.txt'
            with open(done_part_file_path, 'r') as f:
              lines = f.readlines()
              for line in lines:
                line = line.strip()
                if line != '':
                  preds.append(int(line))
        except Exception as e:
            break
        part_i += 1
    return preds
```

```python
def get_metrics(labels,preds):
    if type(preds[0]) == str:
        labels = [str(el) for el in labels]

    labels_clean = []
    preds_clean = []
    for label, pred in list(zip(labels, preds)):
        if pred != 'unpredicted':    
            labels_clean.append(int(label))
            preds_clean.append(int(pred))
    
    return {
        'dataset_len': len(labels),
        'unpredicted_rows_count': preds.count('unpredicted'),
        'filtered_dataset_len': len(labels_clean),
        'accuracy': accuracy_score(labels_clean, preds_clean),
        'precision': precision_score(labels_clean, preds_clean),
        'recall': recall_score(labels_clean, preds_clean)
    }
```

```python
def print_metrics(labels, preds):
    if type(preds[0]) == str:
        labels = [str(el) for el in labels]

    labels_clean = []
    preds_clean = []
    for label, pred in list(zip(labels, preds)):
        if pred != 'unpredicted':    
            labels_clean.append(int(label))
            preds_clean.append(int(pred))
    
    print(f"Dataset length - {len(labels)}")
    print(f"Unpredicted rows - {preds.count('unpredicted')}")
    print(f"Filtered dataset length - {len(labels_clean)}")
    print(f"Accuracy - {accuracy_score(labels_clean, preds_clean)}")
    print(f"Precision - {precision_score(labels_clean, preds_clean)}")
    print(f"Recall - {recall_score(labels_clean, preds_clean)}")
```

```python
def get_frac_k(text):
    text = text.split('_')[-2:]
    
    if len(text[-2]) > 2:
        return text[-1], 4
    else:
        return text[-2], text[-1]

def get_pd_data(datasets_metrics):
    dataset_names = list(datasets_metrics.keys())
    data={
        'dataset_name': [],
        'dataset_len': [],
        'unpredicted_rows_count': [],
        'filtered_dataset_len': [],
        'rag_frac': [],
        'rag_k': [],
        'accuracy': [],
        'precision': [],
        'recall': [],
        'rank': []
    }

    ranks = {
            'twitter': 1,
            'mails': 2,
            'us_news': 3
    }
    
    for dataset_name in dataset_names:
        dataset = datasets_metrics[dataset_name]
        
        for exp_var in list(dataset.keys()):
            exp_metrics = dataset[exp_var]

            rag_frac, rag_k = get_frac_k(exp_var)
            
            data['dataset_name'].append(dataset_name)
            data['rag_frac'].append(int(rag_frac))
            data['rag_k'].append(int(rag_k))
            data['rank'].append(ranks[dataset_name])
            
            for metric_name in list(exp_metrics.keys()):
                metric_value = exp_metrics[metric_name]

                if '.' in str(metric_value):
                    data[metric_name].append(str(metric_value * 100)[:4] + '%')
                else:
                    data[metric_name].append(metric_value)
                
                
    return data
```

```python
def calc_f1_metric(row):
    precision = float(row['precision'].split('%')[0]) / 100
    recall = float(row['recall'].split('%')[0]) / 100

    print(precision + recall)
    return 2 * precision * recall / (precision + recall)

results_df['f1'] = results_df[['precision', 'recall']].apply(lambda row: calc_f1_metric(row), axis=1)
```
