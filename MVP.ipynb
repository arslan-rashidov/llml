```python
import json
import random
import csv
import time

from langchain.schema import HumanMessage, SystemMessage
from langchain.chat_models.gigachat import GigaChat

from read_telegram_data import get_messages_texts_from_json
```

```python
def get_messages_texts_from_json(json_path):
    texts = []

    with open(json_path, encoding='utf-8') as f:
        data = json.load(f)

        for message in data['messages']:
            message_text = message['text']

            text = ''

            if type(message_text) == list:
                for text_raw in message_text:
                    if type(text_raw) == dict:
                        text += ' ' + text_raw['text']
                    else:
                        text += text_raw
            else:
                text = message_text

            texts.append(text)

    return texts
```

```python
texts = get_messages_texts_from_json('taxi-group-messages-feb-4.json')
random_texts = random.sample(texts, 100)
```

```python
labels = []

for text in random_texts:
    print(text)
    
    label = input('Label: ').strip()
    labels.append(label)
```

```python
text_labels = zip(random_texts, labels)
```

```python
unique_texts = {}
for text, label in text_labels:
    if text in list(unique_texts.keys()):
        continue
    else:
        unique_texts[text] = label
```

```python
with open('mvp_texts.csv', 'w', newline='') as csvfile:
    fieldnames = ['text', 'label']
    
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames, delimiter=';')

    writer.writeheader()

    for text, label in unique_texts.items():
        writer.writerow({'text': text, 'label': label})
```

## With RAG

```python
labels_names = {
    '0': 'не спам',
    '1': 'спам'
}
```

```python
texts_for_rag = random.sample(list(zip(texts, labels)), 5)
```

```python
preprocessed_texts_for_rag = ["Это пример класса '" + labels_names[label] + "' - " + text for text, label in texts_for_rag]
```

```python
preprocessed_texts_for_rag
```

```python
from langchain.vectorstores import Chroma
from langchain_community.embeddings import GigaChatEmbeddings

embeddings = GigaChatEmbeddings(
    one_by_one_mode=True, _debug_delay=0.005, credentials='OGIyMjAxOWQtZjYwNi00NzUyLWJlMWYtYmIyMTI0ZDE4MWI0OjUwYjdkYzNmLWU5NDgtNGJjMi04ZWM5LTRkZWQ0MWNlZWI0Ng==', verify_ssl_certs=False
)

vectordb = Chroma.from_texts(preprocessed_texts_for_rag, embeddings)
```

```python
docs = vectordb.similarity_search("пример класса 'не спам'", k=1)
```

```python
docs
```

```python
from langchain.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(chat, retriever=vectordb.as_retriever())
```

```python
prompt = "Напиши 1, если этот текст относится к классу 'спам' или 0, если этот текст относится к классу 'не спам': "
```

```python
rag_preds = []
```

```python
for text in texts:
    query = prompt + text
    result = qa_chain({"query": query})['result']
    #print(result)
    rag_preds.append(result)
    time.sleep(3)
```

```python
true = 0
for label, pred in zip(labels, rag_preds):
    if label == pred:
        true += 1
print(f"Right classificated - {true}/{len(labels)}({true/len(labels)})")
```

```python
for text, label, pred in zip(texts, labels, rag_preds):
    t = text.strip().replace("\n", "")
    print(f'{t}\nLabel-{label}; Pred-{pred}\n----')
```
